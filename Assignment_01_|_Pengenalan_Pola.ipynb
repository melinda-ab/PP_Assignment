Buatlah fungsi untuk melakukan preprocessing dan ekstraksi ciri TF-ID pada data teks tanpa menggunakan library.


```python
import math
import re

def preprocess(text):
  clean_text=re.sub(r'[^a-zA-Z\s]','',text.lower()) #pemotongan & case folding in lexical analysis
  return clean_text.split() #tokenization

def tf(word_dict, bow):
  tf_dict = {}
  bow_count = len(bow) #hitung brp banyak dlm back of word (bow)
  for word, count in word_dict.items():
    tf_dict[word] = count/float(bow_count) #hitung msng-msng word muncul brp kl
  return tf_dict

def idf(doc_list):
  idf_dict = {}
  n = len(doc_list) #hitung total dokumen
  all_words = set([word for doc in doc_list for word in doc]) #bikin database semua kata unik
  for word in all_words:
    jumlah_dokumen_berisi_kata = sum(1 for doc in doc_list if word in doc)
    idf_dict[word] = math.log(n / float(jumlah_dokumen_berisi_kata)) #rumus idf
  return idf_dict

def tf_idf(corpus):
  tokenized_docs = [preprocess(doc) for doc in corpus] #run preprocess

  vocab = sorted(list(set([word for doc in tokenized_docs for word in doc]))) #rapihin & urutin semua kata unik

  word_counts = []
  for doc in tokenized_docs:
      d = dict.fromkeys(vocab, 0)
      for word in doc:
          d[word] += 1
      word_counts.append(d) #data jumlah kata

  tf_list = [tf(word_counts[i], tokenized_docs[i]) for i in range(len(tokenized_docs))] #run tf untuk masing-masing

  idf_list = idf(tokenized_docs) #run idf

  tfidf_matrix = []
  for tf_doc in tf_list:
      hasil_kali = {}
      for word, nilai_tf in tf_doc.items():
          hasil_kali[word] = round(nilai_tf * idf_list[word], 2) #kali masing-masing
      tfidf_matrix.append(hasil_kali)

  return tfidf_matrix, vocab

#test
my_corpus = [
    "Saya suka makan nasi goreng",
    "Nasi goreng itu enak sekali",
    "Saya suka makan sate ayam",
    "Sate ayam itu makanan favorit saya",
    "Masakan ibu sangat enak sekali",
    "Saya beli sate ayam di pasar",
    "Ibu memasak nasi goreng di dapur",
    "Makan sate ayam bersama ibu sangat seru",
    "Ibu membeli bahan masakan di pasar tradisional",
    "Nasi goreng adalah makanan yang sangat populer di Indonesia",
    "Sate ayam dijual di pinggir jalan dekat pasar",
    "Ibu mengajar saya cara memasak nasi goreng yang enak",
    "Makan bersama keluarga besar di pasar malam sangat menyenangkan",
    "Favorit saya adalah makan sate ayam dengan bumbu kacang",
    "Pasar itu sangat ramai karena banyak orang mencari makanan"
]
hasil, vocab = tf_idf(my_corpus)

print("Daftar Kata (Vocab):")
print(vocab)
print()

for i, skor_per_dokumen in enumerate(hasil):
    print(f"Hasil Dokumen {i+1}:")
    baris_angka = [str(skor_per_dokumen[kata]) for kata in vocab]
    print("  ".join(baris_angka))
```
